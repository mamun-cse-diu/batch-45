#Day-1


```python
import pandas as pd

dataset = pd.read_csv('/content/50_Startups.csv')
dataset.head()
```

```python
dataset.info()
```

```python
dataset
```

```python
# imputation -> model based, mean, median, mode
dataset['Marketing Spend'] = dataset['Marketing Spend'].fillna(dataset['Marketing Spend'].mean())
```

```python
dataset.info()
```

```python
dataset['State'] = dataset.State.fillna(dataset.State.mode()[0])
```

```python
dataset.info()
```

```python

from sklearn.preprocessing import LabelEncoder
lb = LabelEncoder()
dataset['State_t'] = lb.fit_transform(dataset.State)
```

```python
dataset
```

```python
import matplotlib.pyplot as plt
plt.scatter(dataset.Profit, dataset['R&D Spend'])
plt.title("Scatter plot")
plt.xlabel("Profit")
plt.ylabel("R&D Spend")
plt.show();
```

#Day-2

```python
data2 = dataset.drop(columns = ['State'])
```

```python
data2
```

```python
x = data2.drop(columns = ['Profit'])
y = data2['Profit']
```

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(
    x,y, random_state=42, test_size= 0.2
)
```

#model training and evaluation



```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(xtrain,ytrain)
lr.score(xtest,ytest)
```

```python
from sklearn.metrics import mean_squared_error as mse
predicted = lr.predict(xtest)
error = mse(ytest, predicted) ** .5
print(error)
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression

models = [RandomForestRegressor(), DecisionTreeRegressor(), LinearRegression()]

errors = {}

for model in models:
    model.fit(xtrain,ytrain)
    name = type(model).__name__
    score = model.score(xtest,ytest)*100
    errors[name] = score
```

```python
errors
```

```python
import matplotlib.pyplot as plt

plt.bar(errors.keys(),errors.values())
```
